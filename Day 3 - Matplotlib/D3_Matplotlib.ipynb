```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
# Functional method for plotting graphs using matplotlib
plt.plot(x,y, 'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('X-Y Graph')
plt.show
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

```python
# Object Oriented for plotting matplotlib
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # Left, Bottom, Width, Height
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('X-Y Graph')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # 20% from left, 50 % from bottom, 40% of canvas size and width, 30% of canvas size and height

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
# Making subplots using OOP
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.tight_layout() # Takes care of overlapping plots
```

```python
# Figure Size and DPI
```

```python
fig = plt.figure(figsize=(8,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1,figsize=(8,4))

# axes[0] = fig.add_axes([0,0,1,1])
axes[0].plot(x,y)

# axes[1] = fig.add_axes([0,0,1,1])
axes[1].plot(y,x)
```

```python
# Adding a legend
```

```python
fig = plt.figure(figsize=(8,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label = 'X_Squared')
ax.plot(x,x**3, label='X_cube')

ax.legend(loc=0)
```

```python
# Plot Appearance
```

```python
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='red',lw=5,alpha=0.5, linestyle='-.') # mlinewidth - 3 thickness, alpha (transparency) - 0.5, linestyle - dot-dash
ax.plot(x,x**3, color='#000000',linewidth=0.5,marker='o', markersize=20, markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green') # RGB Hex code can be used for color
```

```python
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=5,alpha=0.5, linestyle='--')

# Reducing graph size to 'zoom' into area
ax.set_xlim([0,1])
ax.set_ylim([0,1])
```
